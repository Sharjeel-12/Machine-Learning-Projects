### Plotting a entropy function where, 
### $entropy function: -p_1log_2(p_1)-(1-p_1)log_2(1-p_1) $

```python
plt.close()
x=np.arange(0.001,1.001,0.01)
y=-x*np.log2(x)-(1-x)*np.log2(1-x)
plt.plot(x,y,color='blue')
plt.scatter(x,y,c='red',s=8)
plt.xlabel(r'$p_1$')
plt.ylabel('entropy')
plt.title(r'$entropy function: -p_1log_2(p_1)-(1-p_1)log_2(1-p_1) $')
plt.show()
```

```python
X_train[0]
```

### Define a function that finds entropy

```python
def entropy(p):
    if p==0 or p==1:
        H=0
    else:
        H=-p*np.log2(p)-(1-p)*np.log2(1-p)
    return H
```

```python
print(entropy(0.5))
```

```python
def split_indices(X,feature_indices):
    # X contains the training examples
    # feature_indices is the feature on which we want to make the split
    # e.g to split on the "Ear shape" feature spli_indices=0 etc.
    right_splits=[]
    left_splits=[]
    for i,x in enumerate(X):
        if x[feature_indices]==1:
            left_splits.append(i)
        if x[feature_indices]==0:
            right_splits.append(i)
    return left_splits,right_splits
```

### Lets test our function

```python
# 0 means the Ear shape
split_indices(X_train,0)
```

## Lets find the information gain of each split which is given by: -
####  $Information \; gain = H(p_1^{root})-(w^{left}p_1^{left}+w^{right}p_1^{right})$

```python
# this function provides information gain of a feature of index: "feature_indices"
def get_InfoGain(x,y,feature_indices):
    # X contains the training examples
    # feature_indices is the feature on which we want to make the split
    # e.g to split on the "Ear shape" feature spli_indices=0 etc.
    p_root=np.sum(y)/len(y)
    left_splits,right_splits=split_indices(x,feature_indices)
    w_left=len(left_splits)/len(y)
    w_right=len(right_splits)/len(y)
    p_left=np.sum(y[left_splits])/len(left_splits)
    p_right=np.sum(y[right_splits])/len(right_splits)
    return entropy(p_root)-(w_left*entropy(p_left)+w_right*entropy(p_right))
```

## Lets test our function

```python
get_InfoGain(X_train,y_train,2)
```

```python
cat_features=["EarShape","FaceShape","Whiskers"]
IG={}
for i, feature in enumerate(cat_features):
    ig=get_InfoGain(X_train,y_train,i)
    IG[feature]=ig
    
```
