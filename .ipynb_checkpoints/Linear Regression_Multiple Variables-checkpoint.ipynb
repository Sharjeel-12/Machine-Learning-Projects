# Linear Regression for Multiple Variables
### Lets import the relative gear

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating Training Data

```python
x1_train=np.random.uniform(-10,10,(20,))
x2_train=np.random.uniform(-2,2,(20,))
x_train=np.column_stack((x1_train,x2_train))
x_train.shape
```

```python
noise=np.random.uniform(-50,60,(20,))
y_train=3*x1_train+5*x2_train-12+noise
y_train.shape
```

## Function that computes Cost for LR for Muliple Variables

```python
# x is the array of order m by n 
# m is the total ietration number
# n is the number of different features
# b is a scalar
# w is an array of weights of features
def compute_cost(x,y,w,b):
    w=w.reshape(-1,1)
    y=y.reshape(-1,1)
    m=x.shape[0]
    f_wb=np.dot(x,w)+b
    cost=(1/(2*m))*np.sum((f_wb-y)**2)
    return cost
```

```python
compute_cost(x_train,y_train,np.array([1,1]),2)
```

```python
b=100
w_1=np.arange(-1000,1000)
w_2=np.arange(-1000,1000)
```

## Lets check out the variation of the cost with w_1 and w_2 separately

```python
costs_w1=[]
costs_w2=[]
for i in range(w_1.shape[0]):
    costs_w1.append(compute_cost(x_train,y_train,np.array([w_1[i],0]),b))
    costs_w2.append(compute_cost(x_train,y_train,np.array([0,w_2[i]]),b))
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(15,2))
```

```python
ax1.plot(w_1,costs_w1,'r--',lw=5)
ax2.plot(w_2,costs_w2,'g--',lw=5)
plt.show()
```

## Computing the gradients of weights and bias

```python
def compute_gradient(x,y,w,b):
    #dj_dw=np.zeros(w.shape[0])
    dj_db=0
    w = w.reshape(-1,1)
    y = y.reshape(-1, 1)
    m=x.shape[0]
    f_wb=np.dot(x,w)+b
    delta=f_wb-y
    dj_dw=(1/m)*np.dot(x.T,delta)
    dj_db=(1/m)*np.sum(delta)
    return dj_dw,dj_db
```

```python
compute_gradient(x_train,y_train,np.array([1,1]),2)
```

```python
def compute_gradientDescent(x,y,w,b,iterations,alpha):
    J_history=[]
    weight_records=[]
    bias_record=[]
    w=w.reshape(-1,1)
    for i in range(iterations):
        dj_dw,dj_db=compute_gradient(x,y,w,b)
        
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
        J_history.append(compute_cost(x,y,w,b))
        weight_records.append(w)
        bias_record.append(b)
        if (i+1)%100==0:
            print(f'At iteration: {i+1}, the value of the cost function is: {J_history[-1]}')
    return w,b,J_history,weight_records,bias_record
```

```python
w_initial=np.array([1,1])
b_init=10
iterations=2000
alpha=0.01
w_final,b_final,J_history,weight_records,bias_record=compute_gradientDescent(x_train,y_train,w_initial,b_init,iterations,alpha)
```
