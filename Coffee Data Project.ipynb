## Just imported the previous functions 

```python
def g(z):
    return 1/(1+np.exp(-z))
def compute_cost(X,Y,W,b):
    # X is the input matrix of shape (m,n) where m is the number of samples and n=number of features
    # Y is the array of data labels of shape (m,)
    # W is the array of shape (n,)
    m,n=X.shape
    cost=0
    for i in range(m):
        z_wb=0
        for j in range(n):
            z_wb+=W[j]*X[i][j]
        z_wb=z_wb+b
        f_wb=g(z_wb)
        cost+=(-1/m)*(Y[i]*np.log(f_wb)+(1-Y[i])*np.log(1-f_wb))
    return cost

def compute_gradient(X,Y,W,b):
    # X is the input matrix of shape (m,n) where 
    # m = total number of samples and n = number of features
    # Y is the array of data labels of shape (m,)
    # W is the array of shape (n,)
    m,n=X.shape
    dj_dw=np.zeros(n)
    dj_db=0
    for i in range(m):
        z_wb=0
        for j in range(n):
            z_wb+=W[j]*X[i][j]
        z_wb=z_wb+b
        f_wb=g(z_wb)
        delta=f_wb-Y[i]
        dj_dw+=(1/m)*(delta*X[i])
        dj_db+=(1/m)*delta
    return dj_dw,dj_db
def compute_BatchGD(X,Y,W,b,iterations,learning_rate):
    # X is the input matrix of shape (m,n) where 
    # m = total number of samples and n = number of features
    # Y is the array of data labels of shape (m,)
    # W is the array of shape (n,)
    # iterations is the number of epochs for training
    # learning_rate is the speed at which we will be training
    n=W.shape[0]
    J_hist=[]
    alpha=learning_rate
    for i in range(iterations):
        dj_dw,dj_db=compute_gradient(X,Y,W,b)
        for j in range(n):
            W[j]=W[j]-alpha*dj_dw[j]
        b=b-alpha*dj_db
        J_hist.append(compute_cost(X,Y,W,b))
        if (i+1)%100==0:
            print(f'iteration: {i+1}, cost function: {J_hist[-1]}')
    return W,b,J_hist      
```

## Importing the Coffee Data using `pickle`

```python
from sklearn import preprocessing
```

```python
temp=preprocessing.scale(x_train[:,0])
duration=preprocessing.scale(x_train[:,1])
X=np.column_stack((temp,duration,temp**2,temp**3,temp**4,temp**5,temp**6,temp**7,temp**8,temp**9))
X.shape
```

## <span style='color:red;'>So Scaling is very Important</span>

```python
W_init=np.random.uniform(-1,1,(10,))
b_init=1
iterations=3000
learning_rate=0.01
W_final,b_final,J_hist=compute_BatchGD(X,y_train,W_init,b_init,iterations,learning_rate)
```
