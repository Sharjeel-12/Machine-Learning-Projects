# $Normal \;Distribution\; is\; given\; by:\;-$
# $p(x)=\frac{1}{\sqrt{2\pi}\sigma}\;e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$

lets plot this curve

```python
def probability(x,mean,stdev):
    return (1/(np.sqrt(2*np.pi)*stdev))*np.exp(-((x-mean)**2)/(2*(stdev**2)))
```

```python
## testing our probability function
```

```python
x=np.arange(0,10,0.1)
probabilities=probability(x,5,1)
plt.figure()
plt.clf()
plt.plot(x,probabilities,'--',color='blue')
```
